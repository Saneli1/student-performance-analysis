```python
##Find the average math, reading, and writing scores for male and female students separately.
tot_math_male = 0
tot_read_male = 0
tot_write_male = 0
tot_math_fem = 0
tot_read_fem = 0
tot_write_fem = 0
count_male = 0
count_fem = 0
with open(file_path, 'r') as file:
    reader = csv.reader(file)
    next(reader)
    for index, row in enumerate(reader):
        # Convert scores to integers
        math_score = int(row[5])
        reading_score = int(row[6])
        writing_score = int(row[7])
        if row[0] == "male":
            tot_math_male += math_score
            tot_read_male += reading_score
            tot_write_male += writing_score
            count_male += 1
        elif row[0] == "female":
            tot_math_fem += math_score
            tot_read_fem += reading_score
            tot_write_fem += writing_score
            count_fem += 1
    if count_male > 0:
        avg_male_math = tot_math_male / count_male
        avg_male_read = tot_read_male / count_male
        avg_male_write = tot_write_male / count_male
    if count_fem > 0:
        avg_fem_math = tot_math_fem / count_fem
        avg_fem_read = tot_read_fem / count_fem
        avg_fem_write = tot_write_fem / count_fem
print(f"Avg_male_math:{avg_male_math:.2f}")
print(f"Avg_male_read:{avg_male_read:.2f}")
print(f"Avg_male_write:{avg_male_write:.2f}")
print(f"Avg_fem_math:{avg_fem_math:.2f}")
print(f"Avg_fem_math:{avg_fem_math:.2f}")
print(f"Avg_fem_math:{avg_fem_math:.2f}")
```

```python
##Find out if students who took the course performed better on average.
import csv

# Initialize dictionaries
scores = {
    "completed_math": 0, "completed_reading": 0, "completed_writing": 0,
    "none_math": 0, "none_reading": 0, "none_writing": 0
}

counts = {"completed": 0, "none": 0}

# Read the CSV file
with open(file_path, 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header

    for row in reader:  # Loop through each row
        # Convert scores to integers
        math_score = int(row[5])
        reading_score = int(row[6])
        writing_score = int(row[7])
        test_prep = row[4]  # Test preparation status: "completed" or "none"

        # Update the appropriate category
        if test_prep == "completed":
            scores["completed_math"] += math_score
            scores["completed_reading"] += reading_score
            scores["completed_writing"] += writing_score
            counts["completed"] += 1
        elif test_prep == "none":
            scores["none_math"] += math_score
            scores["none_reading"] += reading_score
            scores["none_writing"] += writing_score
            counts["none"] += 1

# Print totals before computing averages
print("Total Scores:", scores)
print("Total Counts:", counts)

# Compute averages after the loop
avg_scores = {}

if counts["completed"] > 0:
    avg_scores["completed_math"] = scores["completed_math"] / counts["completed"]
    avg_scores["completed_reading"] = scores["completed_reading"] / counts["completed"]
    avg_scores["completed_writing"] = scores["completed_writing"] / counts["completed"]

if counts["none"] > 0:
    avg_scores["none_math"] = scores["none_math"] / counts["none"]
    avg_scores["none_reading"] = scores["none_reading"] / counts["none"]
    avg_scores["none_writing"] = scores["none_writing"] / counts["none"]

# Print the final averages
print("Average Scores:", avg_scores)
```

```python
##Find the subject where students scored the highest on average.
# Find highest average for students who completed the test prep course
if avg_scores["completed_math"] > avg_scores["completed_reading"] and avg_scores["completed_math"] > avg_scores["completed_writing"]:
    highest_completed_subject = "Math"
    highest_completed_score = avg_scores["completed_math"]
elif avg_scores["completed_reading"] > avg_scores["completed_writing"]:
    highest_completed_subject = "Reading"
    highest_completed_score = avg_scores["completed_reading"]
else:
    highest_completed_subject = "Writing"
    highest_completed_score = avg_scores["completed_writing"]

print(f"Students who completed the test prep course scored highest on: {highest_completed_subject} ({highest_completed_score:.2f})")

# Find highest average for students who did NOT complete the test prep course
if avg_scores["none_math"] > avg_scores["none_reading"] and avg_scores["none_math"] > avg_scores["none_writing"]:
    highest_none_subject = "Math"
    highest_none_score = avg_scores["none_math"]
elif avg_scores["none_reading"] > avg_scores["none_writing"]:
    highest_none_subject = "Reading"
    highest_none_score = avg_scores["none_reading"]
else:
    highest_none_subject = "Writing"
    highest_none_score = avg_scores["none_writing"]

print(f"Students who did NOT complete the test prep course scored highest on: {highest_none_subject} ({highest_none_score:.2f})")
```

```python
# Identify if there is a trend between parental education level and student performance.

education_counts = {
    "bachelor's degree": 0, "some college": 0, "associate's degree": 0,
    "high school": 0, "some high school": 0, "master's degree": 0
}

education_scores = {
    "bachelor's degree": [0, 0, 0],  
    "some college": [0, 0, 0],
    "associate's degree": [0, 0, 0],
    "high school": [0, 0, 0],
    "some high school": [0, 0, 0],
    "master's degree": [0, 0, 0]
}

# Read the dataset
with open(file_path, 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header

    for row in reader:
        parental_education_level = row[2]
        math_score = int(row[5])
        reading_score = int(row[6])
        writing_score = int(row[7])

        if parental_education_level in education_scores:
            education_scores[parental_education_level][0] += math_score
            education_scores[parental_education_level][1] += reading_score
            education_scores[parental_education_level][2] += writing_score
            education_counts[parental_education_level] += 1  # Count students

# Compute average scores
average_scores = {}
for level in education_scores:
    student_count = education_counts[level]
    
    if student_count > 0:
        avg_math = education_scores[level][0] / student_count
        avg_reading = education_scores[level][1] / student_count
        avg_writing = education_scores[level][2] / student_count
    else:
        avg_math, avg_reading, avg_writing = 0, 0, 0  # Avoid division by zero
    
    average_scores[level] = [avg_math, avg_reading, avg_writing]

# Print results
print("Average Scores by Parental Education Level:")
for level, scores in average_scores.items():
    print(f"{level}: Math: {scores[0]:.2f}, Reading: {scores[1]:.2f}, Writing: {scores[2]:.2f}")
```
